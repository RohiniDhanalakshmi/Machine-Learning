```python
import pandas as pd #pandas is a liberary that use to read dataset file and tabular column. assign pandas as pd 
```

```python
dataset=pd.read_csv("Salary_Data.csv") #the dataset save in csv format (csv-comma separeated value).made pd to read the dataset(Salary_Data.csv).
#give a variable name (dataset) to assgin read dataset.
```

```python
dataset #called the varible to see the input
```

```python
independent=dataset[["Years of experience"]]# splitting the input from dataset substitute respective column name and to assign to independent variable name
independent
```

```python
dependent=dataset[["Salary"]] # splitting output from the dataset with respective column name (salary) to assign with dependent varible name and called the vari
dependent#called the varible name(dependent)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)#splitting into two test and train set.made the model to
#learn with train set and test set to exam the model. sklearn is liberary that contain lots of folder (model_selection) from that folder import function
#(train_test_split)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train) #sklearn is liberary that contain lots folder(linear_model) from that folder import function (LinearRegression) assign
#LingerRegressor to variable name regressor.X_train,y_train pass inside the parameter 
```

```python
y_pred=regressor.predict(X_test) #exam the regressor. made the regressor to predict the X_test answer
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(y_pred,y_test)
r2score # sklearn is liberary the contain lots folder(metrics) from the folder import r2_score that exam the model how it learn and give the result and 
# the model is completely made 
```

```python
import pickle
filename="final_model_SLR.sav" #pickle is liberary that use save and deplyoment the model.here assign a name for the model (final_model_SLR) shouold 
#should save in .sav
```

```python
pickle.dump(regressor,open(filename,"wb")) #here opening the regressor in pickle liberary and the filename 
```

```python
load_model=pickle.load(open("final_model_SLR.sav","rb")) # loding the model ( final_model_SLR..save) and giving the new inuput to predict,to check the
#prediction is correct or not 
result=load_model.predict([[13]])
```
